```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
#Set the hypothesis

#H0: mu_HP Dragon = mu_HP Grass
#H1: mu_HP Dragon != mu_HP Grass

#Significance level= 0.05
```

```python
df["Type 1"].unique()
```

```python
df_dragon = df[df["Type 1"]=="Dragon"]["HP"]
df_grass = df[df["Type 2"]=="Grass"]["HP"]
```

```python
df_dragon.mean(), df_grass.mean()
```

```python
st.ttest_ind(df_dragon,df_grass, equal_var=False)
```

```python
# thence in means between group 1 and group 2 is statistically significant. t = 3.84, p ≈ 0.00032, two-sided. Dragon class has a higher mean than Grass type.
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
df_legendary=df[df['Legendary']==True]
df_notlegendary=df[df['Legendary']==False]
```

```python
df.columns
```

```python
stats_cols= ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
results=[]

for col in stats_cols:
    t, p = st.ttest_ind(df_legendary[col],df_notlegendary[col], equal_var=False)
    results.append({'stats': col, 't_stat': t, 'p_value': p})

results_df=pd.DataFrame(results)
results_df

```

```python
#based on the result we can say that legendary pokemos are superior in every stat possible, by far. for all six stats the differences are statistically significant. this is strong evidence that legendaries have higher stats than non legendaries. 
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df.head()
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 34)
- Hospital coordinates (-122, 37)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
school_lon, school_lat= -118, 34
hospital_lon, hospital_lat= -122, 37

```

```python
import math

eucl_distance=[]

def euclidean_distance(row, ref_lon, ref_lat):
    return np.sqrt((row['longitude'] - ref_lon)**2 +
                   (row['latitude']  - ref_lat)**2)

df['dist_school'] = df.apply(
    lambda row: euclidean_distance(row, school_lon, school_lat),
    axis=1
)

df['dist_hospital'] = df.apply(
    lambda row: euclidean_distance(row, hospital_lon, hospital_lat),
    axis=1
)
  
df['min_dist'] = df[['dist_school', 'dist_hospital']].min(axis=1)
df['close_to_service'] = df['min_dist'] < 0.5
```

```python
df
```

```python
df_closetoservice=df[df['close_to_service']==True]['median_house_value'] 
df_farfromservice=df[df['close_to_service']==False]['median_house_value']
```

```python
st.ttest_ind(df_closetoservice,df_farfromservice, equal_var=False)
```

```python
df_closetoservice
```

```python
print(f'difference of avg of prices regarding closeness to services :', df_closetoservice.mean() - df_farfromservice.mean())
```

```python
#There is very strong evidence that houses that are close to services, have a higher value that the ones that are not. On average, houses close to services cost 66273 more than those that are far from services.  with a p value that is virtually 0. 
```
